### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## Analiss exploratorio

```python
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
df.head()
```

#1. Análisis de facturación



```python
ingresos_por_tienda = df.groupby('Tienda')['Precio'].sum()
ingresos_por_tienda
```

```python
import matplotlib.pyplot as plt

ingresos_por_tienda.plot(kind='bar')
plt.title('Ingreso total por tienda')
plt.ylabel('Ingresos')
plt.xlabel('Tienda')
plt.show()
```

# 2. Ventas por categoría

```python
ventas_categoria = df.groupby(['Tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad')
ventas_categoria.head()
```

```python
for tienda_nombre in df['Tienda'].unique():
    data = ventas_categoria[ventas_categoria['Tienda'] == tienda_nombre]
    plt.figure()
    plt.bar(data['Categoría del Producto'], data['Cantidad'])
    plt.title(f'Ventas por categoría - {tienda_nombre}')
    plt.xticks(rotation=45)
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = df.groupby('Tienda')['Calificación'].mean()
calificacion_promedio
```

```python
calificacion_promedio.plot(kind='bar')
plt.title('Calificación promedio por tienda')
plt.ylabel('Calificación promedio')
plt.xlabel('Tienda')
plt.show()
```

# 4. Productos más y menos vendidos

```python
productos_vendidos = df.groupby(['Tienda', 'Producto']).size().reset_index(name='Ventas')
productos_vendidos.head()
```

```python
for tienda_nombre in df['Tienda'].unique():
    data = productos_vendidos[productos_vendidos['Tienda'] == tienda_nombre]

    mas_vendido = data.sort_values('Ventas', ascending=False).head(1)
    menos_vendido = data.sort_values('Ventas', ascending=True).head(1)

    print(f"\n{tienda_nombre}")
    print("Producto más vendido:")
    print(mas_vendido[['Producto', 'Ventas']])
    print("Producto menos vendido:")
    print(menos_vendido[['Producto', 'Ventas']])
```

```python
for tienda_nombre in df['Tienda'].unique():
    data = productos_vendidos[productos_vendidos['Tienda'] == tienda_nombre]
    top5 = data.sort_values('Ventas', ascending=False).head(5)

    plt.figure()
    plt.bar(top5['Producto'], top5['Ventas'])
    plt.title(f'Productos más vendidos - {tienda_nombre}')
    plt.xticks(rotation=45)
    plt.show()
```

# 5. Envío promedio por tienda

```python
envio_promedio = df.groupby('Tienda')['Costo de envío'].mean()
envio_promedio
```

```python
envio_promedio.plot(kind='bar')
plt.title('Costo de envío promedio por tienda')
plt.ylabel('Costo de envío promedio')
plt.xlabel('Tienda')
plt.show()
```

# EXTRA – Análisis del desempeño geográfico

## Distribución geográfica de las ventas

```python
import matplotlib.pyplot as plt

for tienda_nombre in df['Tienda'].unique():
    data = df[df['Tienda'] == tienda_nombre]

    plt.figure(figsize=(6,4))
    plt.scatter(data['lon'], data['lat'], alpha=0.4)
    plt.title(f'Distribución geográfica de ventas - {tienda_nombre}')
    plt.xlabel('Longitud')
    plt.ylabel('Latitud')
    plt.show()
```
